```python
import json
import pandas as pd


with open('items/globus.json', encoding='utf-8') as f:
    globus = list(map(lambda x: x['data']['brand'], json.load(f)))

with open('items/utkonos.json', encoding='utf-8') as f:
    utkonos = list(map(lambda x: x['data']['brand'], json.load(f)))

df1 = pd.DataFrame(set(globus))
df2 = pd.DataFrame(set(utkonos))
df1[1] = df2[0]
df1.to_excel('brands.xlsx')
```

```python
import pandas as pd
import json

with open('successful_results.json', 'r', encoding='utf-8') as f:
    d = json.load(f)

m = 0
for v in d.values():
    if len(v) > m:
        m = len(v)

df = pd.DataFrame(columns=[i for i in range(m)])
for name, items in d.items():
    items = list(map(lambda x: x['product_name'], items))
    df = df.append(dict(enumerate(items)), ignore_index=True)
```

```python
df.to_excel('merge_test.xlsx')
```
